```python
GCC.rename(columns={'laufkont': 'Status', 'laufzeit': 'Duration', 'moral': 'Credit_history', 'verw': 'Purpose', 'hoehe': 'Amount',
                    'sparkont': 'Savings', 'beszeit': 'Employment_duration', 'rate': 'Installment_rate', 'famges': 'Personal_status_sex', 
                    'buerge': 'Other_debters' , 'wohnzeit' : 'Present_resident', 'verm': 'Property', 'alter': 'Age', 'weitkred': 'Other_installment_plans',
                    'wohn': 'Housing', 'bishkred': 'No_credit', 'beruf': 'Job', 'pers': 'People_liable', 'telef': 'Telephone','gastarb': 'Foreign_worker', 
                    'kredit': 'Credit_risk'}, inplace=True)
```

```python
GCC.nunique()
```

```python
GCC = pd.get_dummies(GCC, columns=['Status', 'Credit_history','Purpose','Savings','Employment_duration','Installment_rate',
                            'Personal_status_sex','Other_debters','Present_resident','Property','Other_installment_plans',
                            'Housing','No_credit','Job','People_liable','Telephone','Foreign_worker'])
```

```python
GCC.head()
```

```python
GCC.shape
```

```python
x=GCC.drop(['Credit_risk'],axis=1)
y=GCC['Credit_risk']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
suraj=lr.predict(x_test)
```

```python
from sklearn import metrics
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, suraj)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, suraj))
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, suraj, average='binary')
```
